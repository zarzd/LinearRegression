```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = datasets.make_regression(n_samples=200, n_features=1, noise=15, random_state=0)
X_ = X.flatten()
sns.scatterplot(x=X_, y=y)
plt.show()
```

```python
class LinearRegression:
    def __init__(self, lr=0.01, n_iter=10000):
        self.lr = lr
        self.n_iter = n_iter
        self.w = None
        self.b = None
    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.w = np.random.rand(num_features)
        self.b = 0
        for i in range(self.n_iter):
            y_pred = np.dot(X, self.w) + self.b
            dw = np.dot(X.T, y_pred - y) / num_samples
            db = np.sum(y_pred - y) / num_samples
            self.w = self.w - self.lr * dw
            self.b = self.b - self.lr * db
    def predict(self, X):
        return np.dot(X, self.w) + self.b
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X,y)
y_pred = linear_regression.predict(X)
```

```python
plt.scatter(X, y, label='Data points')
plt.plot(X_, y_pred , color='red', label='Linear regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
